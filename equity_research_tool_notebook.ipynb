```python
import os
import streamlit as st
import pickle
import time
import langchain
from langchain_groq import ChatGroq
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.qa_with_sources.loading import load_qa_with_sources_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import UnstructuredURLLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
```

```python
os.environ["GROQ_API_KEY"] = "yourAPIKey"
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.7,
    groq_api_key="YourAPIKey"
    
)
```

```python
loaders=UnstructuredURLLoader(urls=["https://www.firstpost.com/world/us-stops-scheduling-visa-appointments-for-foreign-students-amid-plans-of-social-media-screening-13892311.html", "https://www.thehindu.com/news/national/manipur/manipur-mla-new-government-formation-governor-raj-bhawan-imphal-bjp/article69628628.ece"])
data = loaders.load()
print(len(data))
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
docs=text_splitter.split_documents(data)
len(docs)
```

```python
docs[9]
```

```python
embeddings=HuggingFaceEmbeddings()
vectorindex_hfe=FAISS.from_documents(docs,embeddings)
```

```python
file_path="vectorindex.pkl"
with open(file_path, "wb") as f:
    pickle.dump(vectorindex_hfe, f)
```

```python
if os.path.exists(file_path):
    with open(file_path, "rb") as f:
        vectorIndex= pickle.load(f)
```

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorIndex.as_retriever())
chain
```

```python
query="Is visa suspension for students in US permanent?"
langchain.debug = True
chain({"question": query},return_only_outputs=True
)
```
